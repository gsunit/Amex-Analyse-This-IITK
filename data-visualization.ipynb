```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("./data/Training_Data_0611N_cleaned.csv")
df.shape
# df.head()
```

### Finding the most correlated features

```python
plt.figure(figsize=(12,10))
corr_matrix = df.corr()
sn.heatmap(corr_matrix, annot=False, cmap=plt.cm.Reds, xticklabels=True, yticklabels=True)
plt.show()
```

```python
corr_matrix = corr_matrix.abs()
thresholds = [0.98, 0.95, 0.85, 0.75]

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

for threshold in thresholds:
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    to_drop
```
