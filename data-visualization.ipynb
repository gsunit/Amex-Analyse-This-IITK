### Cleaning Code

```python
def fix_column(df, x):
    reset_column = []
    for value in df[x]:
        try:
            value = float(value)
        except:
            value = (value.strip())
            value = (value.replace(",", ""))
            value = (value.replace("(", ""))
            value = (value.replace(")", ""))
        reset_column.append(value)
    df["{}_original".format(x)] = df[x]    
    df[x] = reset_column
    return df
```

```python
df = pd.read_csv("./data/Training_Data_0611N.csv")
```

```python
import numpy as np
df = df.replace(' -   ', np.nan)
df.head()
```

```python
df = fix_column(df, 'pre6m_cust_spend')
df = fix_column(df, 'pre6m_cust_non_disc_amt')
df = fix_column(df, 'pre6m_cust_disc_amt')
df = fix_column(df, 'pre6m_cust_outbound_amt')
df = fix_column(df, 'pre6m_cust_online_amt')
df = fix_column(df, 'pre6m_cust_travel_amt')
df = fix_column(df, 'pre6m_cust_retail_amt')
df = fix_column(df, 'pre6m_total_mc_trs')
df = fix_column(df, 'sum_total_line_amt')
df = fix_column(df, 'cust_max_credit_12m_amt')
df = fix_column(df, 'cust_max_remit_12m_amt')
df = fix_column(df, 'cnsumr_chrg_avg_credit_12m_amt')
df = fix_column(df, 'sow_tot_annual_ext_pmt_amt')
df = fix_column(df, 'sow_tot_revol_bal_amt')
df = fix_column(df, 'sow_tot_trans_bal_amt')
```

```python
df['acq_sub_chn'] = df['acq_sub_chn'].astype('category') 
df['acq_type_grp'] = df['acq_type_grp'].astype('category') 
df['fee_type_grp'] = df['fee_type_grp'].astype('category') 
```

```python
df_cleaned = df[['cust_id', 'profitable_flag', 'count_accts', 'cm_age',
       'flag_top_ed_spender', 'flag_cust_fee_paid_6m', 'pre6m_cust_spend',
       'pre6m_cust_roc_cnt', 'pre6m_cust_non_disc_amt',
       'pre6m_cust_non_disc_cnt', 'pre6m_cust_disc_amt', 'pre6m_cust_disc_cnt',
       'pre6m_cust_outbound_amt', 'pre6m_cust_online_amt',
       'pre6m_cust_online_cnt', 'pre6m_cust_travel_amt',
       'pre6m_cust_travel_cnt', 'pre6m_cust_retail_amt',
       'pre6m_cust_retail_cnt', 'pre6m_cust_myca_active',
       'pre6m_cust_mob_logins', 'pre6m_total_mc_trs', 'acq_sub_chn',
       'acq_type_grp', 'fee_type_grp', 'Cust_tenure', 'pre6m_spend_active_ind',
       'highly_utilized_ind', 'min_pay_ind', 'paid_in_full_ind',
       'sum_total_line_amt', 'direct_debit_ind', 'referrals', 'spillover',
       'self_accts', 'Customer Low Quality indicator', 'cdss_most_rcnt_prob',
       'cust_max_credit_12m_amt', 'cust_max_remit_12m_amt', 'cbr_3_score',
       'cnsumr_chrg_actv_cust_cnt', 'cnsumr_chrg_avg_credit_12m_amt',
       'cnsumr_lend_actv_cust_cnt', 'cnsumr_lend_tot_util_ratio',
       'sow_revol_avg_paydown_pct', 'sow_tot_annual_ext_pmt_amt',
       'sow_tot_revol_bal_amt', 'sow_tot_revol_cnt', 'sow_tot_trans_bal_amt',
       'sow_tot_trans_cnt']]
```
